```python
import pandas as pd

# Load the excel file into a pandas DataFrame
df = pd.read_excel('DataMaze_MoneyBall_Dataset.xls')
# keep only the rows where the "category" column contains the word "Group"
df = df[df['category'].str.contains("Group")]

df_test=pd.read_excel('DataMaze_MoneyBall_Dataset.xls')
# Get a list of columns with 'team2' in their name
cols_to_keep = [col for col in df.columns if 'team2' not in col]
cols_to_keep1 = [col for col in df.columns if 'team1'not  in col]
# Keep only the columns that don't have 'team2' in their name
df1 = df[cols_to_keep]
df2 = df[cols_to_keep1]

# Save the updated DataFrame to a new excel file
df1.to_excel('train_updated_team1.xlsx', index=False)
df2.to_excel('train_updated_team2.xlsx', index=False)
# Get the number of columns
print(len(df.columns))
print(len(df1.columns))
print(len(df2.columns))
```

```python
# Get column names
cols = df2.columns

# Change the column names containing "team2" to "team1"
cols = [col.replace('team2', 'team1') if 'team2' in col else col for col in cols]

# Assign the updated column names to the DataFrame
df2.columns = cols
df2.to_excel('train_updated_team2.xlsx', index=False)
```

```python
# Use the pandas concat function to join the two DataFrames
result = pd.concat([df1, df2])

# Save the result to a new .xlsx file
result.to_excel('result.xlsx', index=False)
print(len(result.columns))
```

```python
# get a list of columns to sum up, excluding the "category" column
sum_columns = [col for col in df.columns ]
# group the data by the teamname column
grouped = df.groupby('team1')

# sum up the columns for each group
result = grouped[sum_columns].sum()

# add the teamname column back to the result dataframe
result['teamname'] = result.index


# reset the index of the result dataframe
result = result.reset_index(drop=True)
result.head()
```

```python
import pandas as pd

# Load the train.csv file
df = pd.read_excel('result1 (1).xlsx')

# Get a list of all the column names
cols = df.columns

# Replace all columns containing "team1" with "team"
df.rename(columns={col: col.replace('team1', 'team') for col in cols if 'team1' in col}, inplace=True)

# Save the modified DataFrame to a new file
df.to_excel('modified_result.xlsx', index=False)
```

```python


# Load the train.csv file
df = pd.read_excel('modified_result.xlsx')

# Calculate the pass completion percentage for each team
df['pass_completion_percentage'] = df['passes completed team'] / df['passes team'] * 100

# Find the team with the highest pass completion percentage
highest_team = df.loc[df['pass_completion_percentage'].idxmax()]['teamname']
highest_percentage = df.loc[df['pass_completion_percentage'].idxmax()]['pass_completion_percentage']

# Print the result
print(f"The team with the highest pass completion percentage is {highest_team} with {highest_percentage:.2f}%")
```

```python
# Find the team with the highest possession
highest_team = df.loc[df['possession team'].idxmax()]['teamname']
highest_possession = df.loc[df['possession team'].idxmax()]['possession team']
# Print the result
print(f"The team with the highest possesion is {highest_team} with {highest_possession} minutes")
```

```python
# Find the team with the highest possession
highest_team = df.loc[df['number of goals team'].idxmax()]['teamname']
highest_goals = df.loc[df['number of goals team'].idxmax()]['number of goals team']
# Print the result
print(f"The team with the highest goals is {highest_team} with {highest_goals} goals")
```

```python
# Calculate the pass completion percentage for each team
df['inbehind_offers_percentage'] = df['inbehind offers to receive team'] / df['total offers to receive team'] * 100

# Find the team with the highest pass completion percentage
highest_team = df.loc[df['inbehind_offers_percentage'].idxmax()]['teamname']
highest_percentage = df.loc[df['inbehind_offers_percentage'].idxmax()]['inbehind_offers_percentage']

# Print the result
print(f"The team with the  highest percentage of offers from inbehind is {highest_team} with {highest_percentage:.2f}%")
```
